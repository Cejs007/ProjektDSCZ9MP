```python
columns1 = pd.read_csv('novy_2014.csv', nrows=0).columns
columns2 = pd.read_csv('novy_2015.csv', nrows=0).columns
columns3 = pd.read_csv('novy_2016.csv', nrows=0).columns
columns4 = pd.read_csv('novy_2017.csv', nrows=0).columns
columns5 = pd.read_csv('novy_2018.csv', nrows=0).columns
```

```python
same_columns = (columns1 == columns2).all() and \
               (columns1 == columns3).all() and \
               (columns1 == columns4).all() and \
               (columns1 == columns5).all()

if same_columns:
    print("Všetky súbory majú rovnaké stĺpce.")
else:
    print("Niektoré súbory majú rôzne stĺpce.")
```

```python
if set(columns1) == set(columns2) == set(columns3) == set(columns4) == set(columns5):
    print("Všetky súbory majú rovnaké stĺpce.")
else:
    print("Názvy stĺpcov sa líšia medzi súbormi.")

# Identifikácia rozdielov
for i, columns in enumerate([columns1, columns2, columns3, columns4, columns5], start=1):
    print(f"Stĺpce v subore {i}: {set(columns)}")
```

```python
import pandas as pd

# Načítanie jednotlivých súborov
df1 = pd.read_csv('novy_2014.csv')
df2 = pd.read_csv('novy_2015.csv')
df3 = pd.read_csv('novy_2016.csv')
df4 = pd.read_csv('novy_2017.csv')
df5 = pd.read_csv('novy_2018.csv')
```

```python
# Spojenie všetkých DataFrame do jedného
df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
```
